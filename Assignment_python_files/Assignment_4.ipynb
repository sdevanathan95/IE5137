The position at time t of a ball dropped with zero initial velocity from a height h is given by

$$ y = h - \frac{gt^{2}}{2} $$

where $g = 9.8 m/s^{2}$. Suppose h = 10 m and the ball is dropped at t = 0.\
&nbsp;&nbsp;&nbsp;&nbsp;(a) Find the sequence of times when the ball passes each half meter.<br> 
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;_(Hint: Create a NumPy array ranging from 10 to 0 in -0.5 increments, then solve the above equation for y.)_ \

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;_(Hint: Given an array a with 20 elements, what are the arrays `a[1:20]` and `a[0:19]`?)_


```python
import math
time =[]
def position(x):
    for i in x:
        t = math.sqrt(((2*10)-(2*i))/9.8)
        #print("at:{0}m the time is {1}s".format(i,round(t,3)))
        time.append(t)
    return time
```

```python
e = np.arange(10,-0.5,-0.5)
e
```

```python
times = position(e)
times = np.array(times).reshape(21,1)
times
```

```python
delt = times[1:20]-times[0:19]
delt = delt.T
```

i.e., with the index attribute storing the coordinate names. Include checks that the two points have the same dimensionality and the same coordinate names. Ensure coordinates are aligned before computing the distance.\
Test your code using the following two points:\
`p = pd.Series({'x':25.5, 'y':3.2, 'z':-0.24})`\
`q = pd.Series({'z':7.3, 'x':-4.62, 'y':-1.4})`\
The result should be:\
31.38830992583067


```python
p = pd.Series({"x":25.5,"y":3.2,"z":-0.24})
q = pd.Series({"z":7.3,"x":-4.62,"y":-1.4})
```

```python
def euclindean(p,q):
    euclideandist = 0
    if p.size == q.size:
        for a in p.index:
            for b in q.index:
                if a==b:
                    euclideandist +=((q[b]-p[a])**2)
        euclideandist = math.sqrt(euclideandist)
        print(euclideandist)
    else:
        print("Dimensions dont match")
```
